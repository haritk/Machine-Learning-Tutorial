```python
def calculate(x):
    return (2**(-1))*np.log(x)    # k = 2
```

```python
x = []
```

```python
for i in range(500):
    x.append(random.randint(1, 10000))
```

```python
y = []
```

```python
for i in range(500):
    y.append(calculate(x[i]))
```
